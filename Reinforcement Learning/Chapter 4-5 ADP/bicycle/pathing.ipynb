```python
import numpy as np

import seaborn
import matplotlib.pyplot as plt

last_iteration = np.load("withoutClipping/last_state_trial_1_with_psi_restriction_True_randomised_state_False_goal_False_test_psiRemoved_tanh_False.npy")

print(np.shape(last_iteration))
```

```python
x_f = last_iteration[:,:,5]
y_f = last_iteration[:,:,6]
omega = last_iteration[:,:,0]
steps = last_iteration[:,:,-1]
col = 0
colors = ['red', 'blue', 'green', 'orange', 'black', 'yellow', 'purple', 'pink', 'olive', 'cyan']

for trajectory_number in range(5):
    if last_iteration[0, trajectory_number, -1] == 0:
        col = (col + 1) % len(colors)
    x = x_f[:, trajectory_number]
    y = y_f[:, trajectory_number]
    plt.plot(x, y, color=colors[col])

plt.savefig("path_no_clipping_with_tanh.pdf")
```

```python
for trajectory_number in range(5):
    if last_iteration[0, trajectory_number, -1] == 0:
        col = (col + 1) % len(colors)
    x = steps[:, trajectory_number]
    y = omega[:, trajectory_number]
    plt.plot(x, y, color=colors[col])
plt.savefig("omega_no_clipping_with_tanh.pdf")
```
